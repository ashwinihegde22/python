```python
class Node:
  def __init__(self, fname = None,lname = None,phone = None,hobby = None, next=None): 
    self.fname = fname
    self.lname = lname
    self.phone = phone
    self.hobby = hobby
    self.next = next

class LinkedList:
  def __init__(self):  
    self.head = None
  
  def insert(self, fname,lname,hobby,phone):
    newNode = Node(fname,lname,hobby,phone)
    if(self.head):
      current = self.head
      while(current.next):
        current = current.next
      current.next = newNode
    else:
      self.head = newNode
  
  def search(self, name):
    current = self.head

    while current != None:
        if current.fname == name:
            current.fname = 'Prutvi'
            return True

        current = current.next
    return False

  def removeDuplicates(self, head):

    if self.head is None or self.head.next is None:
        return head

    hash = set()

    current = head
    hash.add(self.head.fname)

    while current.next is not None:

        if current.next.fname in hash:
            current.next = current.next.next
        else:
            hash.add(current.next.fname)
            current = current.next

    return head

  def reverse(self):
    prev = None
    current = self.head
    while(current is not None):
        next = current.next
        current.next = prev
        prev = current
        current = next
    self.head = prev

  def printLL(self):
    current = self.head
    while(current):
      print(current.fname,end=' ')
      print(current.lname,end=' ')
      print(current.hobby,end=' ')
      print(current.phone)
      
      
      current = current.next

LL = LinkedList()
#insert new element
LL.insert('ashwini','h',23456789,'Singing')
LL.insert('ashwini','h',23456789,'Singing')
LL.insert('Sanjay','mk',10000000,'dancing')
print("Data present in LL is : ")
LL.printLL()
print('\n')

#search & rename
if LL.search('Sanjay'):
    print("Found and updated, New Linked list is : ")
    LL.printLL()
else:
    print("Not found")
print('\n')

LL.removeDuplicates(LL.head)
print("LL after removing duplicates")
LL.printLL()
print('\n')


LL.reverse()
print("LL after reversing :- ")
LL.printLL()
print('\n')
```
